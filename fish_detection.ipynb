```python
def plot_movement_me(data, frm, to):
    """
    Plot 'x' and 'y' attributes for given Pandas DataFrame in specified time frame.

    :param data: Pandas DataFrame (should be sorted by 'time' attribute).
    :param frm: Starting from time step.
    :param to: Ending to time step.
    :return: None.
    """

    plt.scatter(x='x',
                y='y',
                data=data.loc[(data['time'] == frm) & (data['average_speed'] < to), :], c='r')
    plt.scatter(x='x',
                y='y',
                data=data.loc[(data['time'] == frm) & (data['average_speed'] >= to), :], c='b')

    
    
#     plt.scatter(x='x',
#                 y='y',
#                 data=data.loc[(data['time'] == frm) & (data['animal_id'] == to), :], c='r')
#     plt.scatter(x='x',
#                 y='y',
#                 data=data.loc[(data['time'] == frm) & (data['animal_id'] != to), :], c='b')


    
    plt.title("'multiple' fishes coordinates")
    plt.xlim(0, 1000)
    plt.ylim(0, 1000)
    plt.xlabel("'x' coordinate")
    plt.ylabel("'y' coordinate")
#     plt.grid()
#     plt.savefig('fig/fig_'+str(frm)+'.png')
    plt.show()

    return None
```

```python
steps =1000

for s in range(1, steps+1, 100):
    plot_movement_me(data_features, frm=s, to=0.1)
    
```
